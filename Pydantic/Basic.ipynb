```python
from pydantic import BaseModel
```

```python
class employee(BaseModel):
    name : str
    id : int
    salary : int
    department : str
```

```python
obj = employee(name = "John", id = 123, salary = 50000, department = "IT")
print(obj)
```

### Coercion VS Strict Types

#### Coercion

```python
obj = employee(**{"name": "John", "id": "123", "salary": "50000", "department": "HR"})
obj
```

```python
obj = employee(name = "John", id = "abc", salary = "50000", department = "IT") 
# raises error because id should be int and pydantic will try it best effort to convert it but it will fail because "abc" cannot be converted to int
```

#### Strict Types


```python
from pydantic import StrictInt

class employee_schema(BaseModel):
    name : str
    id : StrictInt
    salary : int
    department : str
```

```python
obj1  = employee_schema(name = "John", id = "123", salary = "50000", department = "IT")
# raises error because id should be StrictInt and pydantic will not try to convert it because we have used StrictInt which means it should be int and if we pass string it will raise error without trying to convert it
```

```python
obj_2 = employee_schema(**{"name": "John", "id": 123, "salary": 50000, "department": "IT"}) 
# this is another way to create an object using dictionary unpacking(**) and it will work fine because the values are of correct type
obj_2
```
